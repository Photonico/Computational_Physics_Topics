## Kepler's problem

The Kepler problem deals with the motion of an object in a central force field, such as the gravitational pull of the Sun. The total kinetic energy of the object consists of radial and angular components, while its potential energy is governed by gravitational attraction. In central force problems, the effective potential energy combines the actual potential energy with the centrifugal potential energy.

The Kepler problem deals with the motion of an object in a central force field, for example, the gravitational field of the sun, the total energy can be written as:

$$ E = T + V = \frac{\mu}{2}(\dot{r}^2 + r^2\dot{\theta}^2) + \int_r^{\infty}\frac{{\rm G}Mm}{r^2}{\rm d}r   $$

Where, 
$m$ and $M$ are the masses of the two objects;

$\mu$ is the reduced mass: $\mu = \frac{mM}{m+M}$;

${\rm G}$ is the gravitational constant;

We introduce the angular momentum $l$, and we can infer:

$$ E = \frac{\mu}{2}\dot{r}^2 + \frac{l^2}{2\mu r^2} - \frac{{\rm G}Mm}{r} $$

where the term of $\frac{\mu}{2}\dot{r}^2$ denotes the radial kinetic energy $T_r$ of this system:

$$ T_r = \frac{\mu}{2}\dot{r}^2 $$

The total kinetic energy of the object consists of radial and angular components, while its potential energy is governed by gravitational attraction. In central force problems, the effective potential energy combines the actual potential energy with the centrifugal potential energy. 

In Kepler's problem scenario, the actual potential $V(r)$ denotes $-\frac{{\rm G}Mm}{r}$ which represents the gravitational attraction between the objects.

In addition, the centrifugal potential energy $V_c$ denotes $\frac{l^2}{2\mu r^2}$. It is associated with the rotational motion of the object around the center.

Therefore, we can represent the effective potential energy $V_{eff}(r)$:

$$\begin{align*}
V_{eff}(r) &= V_c +  V(r)\\
&= \frac{l^2}{2\mu r^2}-\frac{{\rm G}Mm}{r}
\end{align*}$$

We use a constant $k$ instead of ${\rm G}Mm$, thus, we arrive:

$$V_{eff}(r)= \frac{l^2}{2\mu r^2}-\frac{k}{r}$$

We can use the effective potential energy to simplify a two-dimensional problem into a one-dimensional problem according to the radial coordinate $r$.

Analyzing the effective potential energy provides us with insights into the stability, energy, and shapes of orbits.  It also facilitates a deeper understanding of celestial mechanics governed by Kepler's laws.


```python
## Effective potential energy for two bodies problem

# Packages invoking
import numpy as np; 
import matplotlib.pyplot as plt
from mpl_toolkits.axisartist.axislines import AxesZero


# Computational parameters
step = 0.01
r = np.arange(step, 10, step)

# Physical parameters
k = 2
l = 1
μ = 1

# Centrifugal potential V_c
V_c = np.where(r!=0, (l**2)/(2*μ*r**2),np.nan)
# Actual potential V_r
V_r = np.where(r!=0, -k/r,np.nan)

# Effective potential V_eff
V_e = V_c + V_r

# Matplotlib style parameters
v_lim = np.max(np.abs(V_e))*0.001
x_lim = (0, np.max(r)*0.25)
y_lim = (-v_lim, v_lim)
label = 14
params = {"text.usetex": False, "font.family": "serif", "mathtext.fontset": "cm",
          "axes.titlesize": label+2, "axes.labelsize": label, "figure.facecolor": "w"}
plt.rcParams.update(params)
fig = plt.figure(figsize=(6.4, 4.8), dpi=256)
ax = fig.add_subplot(axes_class=AxesZero)

# Customize the coordinate frame
for direction in ["xzero", "yzero"]:
    ax.axis[direction].set_axisline_style("-|>")
    ax.axis[direction].set_visible(True)
for direction in ["left", "right", "bottom", "top"]:
    ax.axis[direction].set_visible(False)
plt.text(0, 0, r"$(0,0)$", fontsize=label, ha ='left', va='top')
plt.text( np.max(x_lim)*0.975, 0, r"$r$", fontsize=label, ha ='left', va='top')
plt.text(-np.max(x_lim)*0.025, np.max(y_lim), r"$V$", fontsize=label, ha ='right', va='top')

# Graph parameters
plt.title("Effective Potential Energy")
plt.xlim(x_lim)
plt.ylim(y_lim)
plt.xticks([])
plt.yticks([])

# Plot the data
plt.plot(r, V_c, color="#EB731E", label=r"Centrifugal potential")
plt.plot(r, V_r, color="#0078E1", label=r"Actual potential")
plt.plot(r, V_e, color="#AF5AE1", label=r"Effective potential")
plt.legend()

print()
```
